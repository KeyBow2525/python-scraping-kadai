```python
import requests
import pprint
from getpass import getpass

api_key = getpass('APIキーを入力してください:')

search_station = input('検索する駅を入力してください: ')

url_station = 'https://maps.googleapis.com/maps/api/geocode/json'

response_station = requests.get(
    url_station, params = { 
        'address': search_station,
        'key': api_key
    }
)

json_station = response_station.json()
pprint.pprint(json_station)
```

```python
for item in json_station['results']:
    location_lat = item['geometry']['location']['lat']
    location_lng = item['geometry']['location']['lng']

print(f'緯度: {location_lat}，経度: {location_lng}')
```

```python
url_restaurant = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

response_restaurant = requests.get(
    url_restaurant, params = { 
        'location': f'{location_lat},{location_lng}',
        'language': 'ja',
        'radius': 500,
        'type': 'restaurant',
        'key': api_key
    }
)

json_restaurant = response_restaurant.json()
pprint.pprint(json_restaurant)
```

```python
for item in json_restaurant['results']:
    restaurant_name = item['name']
    restaurant_rate = item['rating']
    restaurant_place = item['vicinity']
    print(f'{restaurant_name} {restaurant_rate} {restaurant_place}')
#  {restaurant_rating}
```
